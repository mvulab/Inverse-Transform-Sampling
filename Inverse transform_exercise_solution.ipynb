```python
import random
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
```

A random variable has a Laplace $(\mu, b)$ distribution if its probability density function is
$$
p(x | \mu, b)=\frac{1}{2 b} \exp \left(-\frac{|x-\mu|}{b}\right)
$$

$$F^{-1}(u)=\mu-b \operatorname{sgn}(u-0.5) \ln (1-2|u-0.5|)$$

Generate 5000 samples from this distribution using the inverse transform method. 

Compare the Laplace pdf with the histogram obtained from samples generated with the inverse method
Consider $-10 \leq x \leq 10$ and 100 bins.

```python
def inverse_transform_laplace(b, mu, num_samples):
    u = uniform.rvs(size=num_samples)
    x = mu-b*np.sign(u-0.5)*np.log(1-2*np.abs(u-0.5))
    return x
```

```python
from scipy.stats import laplace

plt.figure(figsize=(8, 4))
plt.hist(inverse_transform_laplace(1, 0, 5000),bins=100, density=True, label='Samples with \ninverse method');
x_n = np.linspace(-10, 10, 100)
plt.plot(x_n, laplace().pdf(x_n), lw=4, label='Laplace pdf')
plt.legend()
plt.xlabel('x')
plt.ylabel('Density')
```
